```python
import pandas as pd

compras = pd.read_csv("compras.csv", index_col="id") #id como indice
compras
```

```python
personas = pd.read_csv("persona.csv")
personas
```

```python
#Contar el número de compras por persona y obtener el top 3
#quiero saber cuantas veces aparece el id personas value count
compras["id_persona"].value_counts().head(3)
```

```python
#Calcular el nombre de la persona que más gastó
#tendriamos que hacer un join
data = pd.merge(compras, personas, left_on="id_persona", right_on="id")
data 
```

```python
data.groupby("nombre")["precio"].sum().sort_values(ascending=False)
```

```python
data.groupby("nombre")["precio"].sum().sort_values(ascending=False).head(1)
```
