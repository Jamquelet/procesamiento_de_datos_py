```python
# 3. Graficar la distribución de aprobados con un pie chart
aprobados = df['aprobado'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(aprobados, labels=aprobados.index, autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
plt.title('Proporción de Estudiantes Aprobados y No Aprobados')
plt.show()
""" Graficar la distribución de aprobados con un pie chart: Creamos un pie chart utilizando Matplotlib para mostrar la proporción de estudiantes aprobados y no aprobados en el dataset.  """ 
```
